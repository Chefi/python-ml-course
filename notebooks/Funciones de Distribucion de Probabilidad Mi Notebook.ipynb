## Simulación de Monte Carlo

Genera 2 aleatorios entre $0$ y $1$: $x$ y $y$.

Calcula $x^2 + y^2 = r^2$

* Si el valor de $r^2$  es menor o igual a $1$, entonces estamos en el interior del círculo de radio $r = 1$

* Si el valor de $r^2$ es mayor o gual a $1$ no estamos en el interior del círculo de radio $r = 1$

Cuenta casos posibles/casos totales y eso estima $\frac{\pi}{4}$. Con diferentes réplicas obtiene aproximacioens de  $\frac{\pi}{4}$

```python
pi_avg = 0
n = 10000
pi_value_list = []
for i in range(100):
    value = 0
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()
    for j in range(n):
        z = x[j]*x[j] + y[j]*y[j]
        if z <= 1:
            value += 1
    float_value = float(value)
    pi_value = float_value*4/n
    pi_value_list.append(pi_value)
    pi_avg += pi_value

pi = pi_avg/100
print(pi)
plt.plot(pi_value_list)
    
```

```python
def pi_montecarlo(npuntos, nexperimentos):
    pi_avg = 0
    pi_value_list = []
    for i in range(nexperimentos):
        value = 0
        x = np.random.uniform(0,1,npuntos).tolist()
        y = np.random.uniform(0,1,npuntos).tolist()
        for j in range(npuntos):
            z = x[j]*x[j] + y[j]*y[j]
            if z <= 1:
                value += 1
        float_value = float(value)
        pi_value = float_value*4/npuntos
        pi_value_list.append(pi_value)
        pi_avg += pi_value
    
    pi = pi_avg/nexperimentos
    print(pi)
    fig = plt.plot(pi_value_list)
    return (pi, fig)
```

```python
pi_montecarlo(npuntos=100000, nexperimentos=500)
```
